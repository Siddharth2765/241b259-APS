# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP tables for costs and split points
    m = [[0 for _ in range(n)] for _ in range(n)]
    s = [[0 for _ in range(n)] for _ in range(n)] # s[i][j] stores the split point k

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                # cost = m[i][k] + m[k+1][j] + (p[i-1] * p[k] * p[j])
                # This is the cost to multiply matrices A_i...A_k and A_{k+1}...A_j
                # plus the cost to multiply the two resulting matrices.
                cost = m[i][k] + \
                       m[k+1][j] + \
                       p[i-1] * p[k] * p[j]

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k  # Store the split point

    return m, s
```

```python
# 1. Given p = [5, 10, 3, 12, 5, 50, 6]
p_exercise = [5, 10, 3, 12, 5, 50, 6]

m_exercise, s_exercise = matrix_chain_multiplication(p_exercise)

# 2. Compute minimum cost
min_cost_exercise = m_exercise[1][len(p_exercise) - 1]
print(f"Minimum multiplication cost for p_exercise: {min_cost_exercise}")
```

```python
# 3. Print optimal parenthesization
def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        k = s[i][j]
        left = print_optimal_parenthesization(s, i, k)
        right = print_optimal_parenthesization(s, k + 1, j)
        return f"({left} x {right})"


print("Optimal Parenthesization:")
print(print_optimal_parenthesization(s_exercise, 1, len(p_exercise) - 1))
```

```python
# 4. Compare number of multiplications saved

def calculate_naive_cost_left_to_right(p_dims):
    if len(p_dims) < 3:
        return 0

    total_cost = 0
    # First multiplication: A1 x A2
    current_rows = p_dims[0]
    current_cols = p_dims[2]
    total_cost += p_dims[0] * p_dims[1] * p_dims[2]

    # Subsequent multiplications: (A1A2) x A3, then (A1A2A3) x A4, etc.
    for i in range(3, len(p_dims)):
        # We are multiplying the accumulated product (current_rows x current_cols)
        # with the next matrix Ai (p_dims[i-1] x p_dims[i])
        # The next matrix dimensions are current_cols x p_dims[i]
        total_cost += current_rows * current_cols * p_dims[i]
        current_cols = p_dims[i]

    return total_cost


naive_cost_exercise = calculate_naive_cost_left_to_right(p_exercise)
saved_multiplications = naive_cost_exercise - min_cost_exercise

print(f"Naive multiplication cost (left-to-right): {naive_cost_exercise}")
print(f"Number of multiplications saved: {saved_multiplications}")
```
